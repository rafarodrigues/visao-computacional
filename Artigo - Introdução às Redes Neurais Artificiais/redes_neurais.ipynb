## 2.1 C√°lculos l√≥gicos com neur√¥nios

Warren McCulloch e Walter Pitts propuseram um modelo simplificado de neur√¥nio biol√≥gico, conhecido como **neur√¥nio artificial**, que possui uma ou mais entradas bin√°rias (ligado/desligado) e uma sa√≠da bin√°ria.

O neur√¥nio artificial ativa sua sa√≠da quando mais de um certo n√∫mero de suas entradas est√£o ativas. Eles demonstraram que, mesmo com um modelo t√£o simplificado, √© poss√≠vel construir uma rede de neur√¥nios artificiais capaz de computar qualquer proposi√ß√£o l√≥gica desejada, assumindo que um neur√¥nio √© ativado quando pelo menos duas de suas entradas est√£o ativas.


image.png

- A primeira rede √† esquerda representa a fun√ß√£o **identidade**: se o neur√¥nio A est√° ativado, o neur√¥nio C tamb√©m ser√° ativado (recebendo dois sinais de entrada do neur√¥nio A), mas se o neur√¥nio A est√° desativado, o neur√¥nio C tamb√©m estar√°. </br></br>

- A segunda rede executa um **AND** l√≥gico: o neur√¥nio C √© ativado somente quando ambos os neur√¥nios A e B est√£o ativados (um √∫nico sinal de entrada n√£o √© suficiente para ativar o neur√¥nio C).</br></br>

- A terceira rede realiza um **OR** l√≥gico: o neur√¥nio C √© ativado se o neur√¥nio A ou B estiver ativado (ou ambos).</br></br>

- Finalmente, supondo que uma conex√£o de entrada possa inibir a atividade do neur√¥nio (o que acontece com neur√¥nios biol√≥gicos), a quarta rede calcula uma proposi√ß√£o **l√≥gica um pouco mais complexa**: o neur√¥nio C √© ativado apenas se o neur√¥nio A estiver ativo e o neur√¥nio B estiver desativado. Se o neur√¥nio A estiver ativo o tempo todo, isso resulta em um NOT l√≥gico: o neur√¥nio C est√° ativo quando o neur√¥nio B est√° desativado, e vice-versa.

## 2.2 Perceptron


O **Perceptron** √© uma das arquiteturas de RNA mais simples, inventada em 1957 por Frank Rosenblatt. Baseia-se em um neur√¥nio artificial um pouco diferente, chamado de *threshold logic unit* (**TLU**) ou, √†s vezes, unidade de *linear threshold unit* (**LTU**), onde as entradas e a sa√≠da s√£o agora n√∫meros (em vez de valores bin√°rios ligado/desligado) e cada conex√£o de entrada est√° associada a um peso.

O TLU calcula uma soma ponderada de suas entradas `(z = w1 x1 + w2 x2 + ‚ãØ + wn xn = xT w)`, aplica uma fun√ß√£o degrau a essa soma e produz o resultado: `hw(x) = step(z), onde z = xTw`.


image.png

Um √∫nico TLU pode ser usado para **classifica√ß√£o bin√°ria** linear simples. Ele computa uma combina√ß√£o linear das entradas e, se o resultado exceder um limiar, produz a classe positiva, ou ent√£o produz a classe negativa. Por exemplo, um √∫nico TLU poderia classificar flores de √≠ris com base no comprimento e largura da p√©tala. Treinar um TLU significa encontrar os valores corretos para **w0, w1 e w2**.

Um Perceptron √© composto simplesmente por uma √∫nica camada de TLUs, com cada TLU conectado a todas as entradas. Quando todos os neur√¥nios em uma camada est√£o conectados a cada neur√¥nio na camada anterior, ela √© chamada de camada totalmente conectada ou **camada densa**.

Para representar que cada entrada √© enviada a cada TLU, √© comum desenhar neur√¥nios de passagem chamados neur√¥nios de entrada, que apenas produzem qualquer entrada que recebam. Todos os neur√¥nios de entrada formam a camada de entrada.

Al√©m disso, geralmente √© adicionada uma camada de *bias feature* `x0 = 1`, representada por um tipo especial de neur√¥nio chamado neur√¥nio de vi√©s, que apenas produz 1 o tempo todo.

Um Perceptron com duas entradas e tr√™s sa√≠das pode classificar inst√¢ncias simultaneamente em tr√™s diferentes classes bin√°rias, o que o torna um classificador de m√∫ltiplas sa√≠das.

Como um Perceptron √© treinado? O algoritmo de treinamento do Perceptron, proposto por Frank Rosenblatt, foi inspirado na regra de Hebb. Donald Hebb sugeriu que, **quando um neur√¥nio biol√≥gico frequentemente ativa outro, a conex√£o entre eles se fortalece**.

Essa ideia foi resumida na frase de Siegrid L√∂wel: "C√©lulas que disparam juntas, se conectam juntas." Conhecida como regra de Hebb, ou aprendizado Hebbiano, o peso da conex√£o entre dois neur√¥nios aumenta quando t√™m o mesmo resultado. Os Perceptrons s√£o treinados usando uma variante dessa regra que considera o erro da rede, refor√ßando conex√µes que ajudam a reduzir o erro.

Especificamente, o Perceptron recebe uma inst√¢ncia de treino de cada vez, fazendo suas previs√µes e, para cada neur√¥nio de sa√≠da com previs√£o errada, refor√ßa os pesos das conex√µes dos inputs que contribuiriam para a previs√£o correta. A fronteira de decis√£o de cada neur√¥nio de sa√≠da √© linear, ent√£o os Perceptrons s√£o incapazes de aprender padr√µes complexos.

## 2.3 Aplica√ß√£o do Perceptron

O Scikit-Learn fornece uma classe Perceptron que implementa uma √∫nica rede TLU. Pode ser usada por exemplo no conjunto de dados iris.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

iris = load_iris()

# Selecionar caracter√≠sticas para uso (comprimento da p√©tala e largura da p√©tala)
X = iris.data[:, (2, 3)]
y = (iris.target == 0).astype(np.int64) # Iris Setosa?

per_clf = Perceptron()

# Armazenar a precis√£o do modelo durante o treinamento
# accuracies = []

# Treinar o modelo por 100 √©pocas
for epoch in range(100):
    per_clf.partial_fit(X, y, classes=np.unique(y))
    y_pred = per_clf.predict(X)
    # accuracies.append(accuracy_score(y, y_pred))


# y_pred = per_clf.predict([[2, 0.5]])
```

```python
# Visualizar os dados e a linha de decis√£o aprendida pelo perceptron
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('Separa√ß√£o de Classes pelo Perceptron')
plt.xlabel('Comprimento da P√©tala')
plt.ylabel('Largura da P√©tala')

# Obter os limites da plotagem
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

# Criar um meshgrid para plotar a linha de decis√£o
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = per_clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Plotar a linha de decis√£o
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)

plt.show()
```

Diferente da Regress√£o Log√≠stica, Perceptrons decidem com base em um limiar fixo, n√£o estimando probabilidades de classe.

Cr√≠ticas de Minsky e Papert em 1969 sobre limita√ß√µes dos Perceptrons, como a incapacidade de resolver o problema XOR, levaram a um desinteresse pela pesquisa em redes neurais.

Contudo, empilhando Perceptrons para formar um Perceptron Multicamadas (MLP), essas limita√ß√µes podem ser superadas, possibilitando resolver o problema XOR e reavivando o interesse em redes neurais.

# 3. Multi-Layer Perceptron e Backpropagation

Uma **MLP** (multi-layer perceptron) √© composto de um *input layer*, uma ou mais camadas de TLU, chamadas de ***hidden layers***, e uma camada final de TLU, chamada ***output layer***.

As camadas pr√≥ximas √† camada de entrada s√£o geralmente chamadas de camadas inferiores, e as pr√≥ximas √†s sa√≠das s√£o chamadas de camadas superiores. Todas as camadas, exceto a camada de sa√≠da, incluem um neur√¥nio de vi√©s e s√£o totalmente conectadas √† pr√≥xima camada.

image.png

Quando uma RNA cont√©m uma pilha profunda de camadas ocultas, ela √© chamada de **rede neural profunda** (DNN).

Em 1986, David Rumelhart, Geoffrey Hinton e Ronald Williams publicaram um artigo inovador introduzindo o algoritmo de treinamento de **retropropaga√ß√£o**, que ainda √© usado hoje.

Em resumo, √© simplesmente o **Gradiente Descendente** usando uma t√©cnica eficiente para calcular os gradientes automaticamente:
> em apenas duas passagens pela rede (uma para frente, uma para tr√°s), o algoritmo de retropropaga√ß√£o consegue calcular o gradiente do erro da rede em rela√ß√£o a cada par√¢metro do modelo.

Em outras palavras, ele pode descobrir como cada peso de conex√£o e cada termo de vi√©s devem ser ajustados para reduzir o erro.

Uma vez que tem esses gradientes, realiza um passo de Gradiente Descendente regular, e todo o **processo √© repetido** at√© a rede convergir para a solu√ß√£o.


image.png

Descrevendo um pouco o algoritimo

1. **Processamento de Mini-lotes**: O algoritmo processa um mini-lote de dados por vez (ex.: 32 inst√¢ncias) e repete o processo para todo o conjunto de treinamento v√°rias vezes, sendo cada repeti√ß√£o denominada uma √©poca.

2. **Forward Pass**: Cada mini-lote √© inicialmente inserido na camada de entrada, passando sequencialmente por cada camada oculta at√© a camada de sa√≠da. Durante esse processo, as sa√≠das de todas as camadas s√£o calculadas e armazenadas para uso posterior.

3. **C√°lculo do Erro**: Ap√≥s a obten√ß√£o da sa√≠da final, o algoritmo calcula o erro de sa√≠da utilizando uma fun√ß√£o de perda, que compara a sa√≠da desejada com a sa√≠da real da rede.

4. **Backward Pass (Retropropaga√ß√£o do Erro)**: Utilizando a regra da cadeia, uma ferramenta fundamental do c√°lculo, o algoritmo retrocede pela rede, camada por camada, calculando quanto cada conex√£o contribuiu para o erro total. Este passo determina o gradiente de erro para todos os pesos da rede.

5. **Gradiente Descendente**: Com os gradientes de erro calculados, o algoritmo ajusta os pesos de todas as conex√µes na rede, realizando um passo de Gradiente Descendente para minimizar o erro.

Este procedimento, repetido para cada inst√¢ncia de treinamento, envolve fazer uma previs√£o (forward pass), medir o erro, calcular a contribui√ß√£o do erro por cada conex√£o (backward pass) e ajustar os pesos para reduzir o erro.


image.png

‚ö†Ô∏è **Procurar sobre Gradient Descent**

**Resumindo**: para cada inst√¢ncia de treinamento, o algoritmo de retropropaga√ß√£o primeiro faz uma previs√£o (passagem direta), mede o erro, em seguida, passa por cada camada em reverso para medir a contribui√ß√£o do erro de cada conex√£o (passagem reversa), e finalmente ajusta levemente os pesos das conex√µes para reduzir o erro (passo do Gradiente Descendente).

```python
# Gerando √≠ndices aleat√≥rios com pelo menos 2 de cada categoria
indices = []
for class_idx in range(10):  # Para cada classe
    valid_indices = np.where(y_train == class_idx)[0]
    indices.extend(np.random.choice(valid_indices, 2, replace=False))

# Adicionando √≠ndices aleat√≥rios at√© completar 24 imagens, garantindo diversidade
while len(indices) < 24:
    indices.append(np.random.randint(0, len(X_train)))

np.random.shuffle(indices)  # Misturando para diversidade visual
indices = indices[:24]  # Garantindo exatamente 24 imagens

# Plotando as imagens
plt.figure(figsize=(10, 7))
for i, idx in enumerate(indices):
    plt.subplot(4, 6, i + 1)
    plt.imshow(X_train[idx], cmap="binary")
    plt.title(class_names[y_train[idx]])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Modelo de classifica√ß√£o MLP com dois hidden layers:
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28, 28]))
model.add(keras.layers.Dense(300, activation="relu"))
model.add(keras.layers.Dense(100, activation="relu"))
model.add(keras.layers.Dense(10, activation="softmax"))
```

# 6 Camadas, hiperpar√¢metros e feature extraction

Al√©m de `Flatten` e `Dense`, voc√™ pode utilizar diversas outras camadas em seus modelos de redes neurais com o Keras, incluindo:

- `Conv2D`: Camada de convolu√ß√£o para imagens, √∫til para modelos de vis√£o computacional.
  - Hiperpar√¢metros importantes: `filters`, `kernel_size`, `strides`, `padding`, `activation`.

- `MaxPooling2D`: Camada de pooling para reduzir a dimensionalidade espacial.
  - Hiperpar√¢metros importantes: `pool_size`, `strides`, `padding`.

- `Dropout`: Camada para reduzir o overfitting ao "desligar" aleatoriamente neur√¥nios durante o treinamento.
  - Hiperpar√¢metros importantes: `rate`.

- `BatchNormalization`: Camada para normalizar as entradas de cada mini-batch, ajudando na estabiliza√ß√£o e acelera√ß√£o do treinamento.
  - Hiperpar√¢metros importantes: `momentum`, `epsilon`.

- `LSTM`: Para modelos de sequ√™ncia, como processamento de linguagem natural.
  - Hiperpar√¢metros importantes: `units`, `activation`, `recurrent_activation`.

- `Embedding`: Usado para mapear representa√ß√µes de entrada inteiras (como √≠ndices de palavras) para embeddings densos.
  - Hiperpar√¢metros importantes: `input_dim`, `output_dim`.

- `SimpleRNN`: Uma vers√£o mais simples e b√°sica de uma camada recorrente para processamento de sequ√™ncias.
  - Hiperpar√¢metros importantes: `units`, `activation`.

- `GRU`: Semelhante ao LSTM, mas mais eficiente em termos computacionais e com menos par√¢metros.
  - Hiperpar√¢metros importantes: `units`, `activation`, `recurrent_activation`.

Para cada uma dessas camadas, voc√™ pode ajustar diversos hiperpar√¢metros para afinar o desempenho do seu modelo. Alguns hiperpar√¢metros comuns incluem:

- `activation`: A fun√ß√£o de ativa√ß√£o a ser usada. Exemplos comuns incluem `"relu"`, `"sigmoid"` e `"tanh"`.
- `units`: O n√∫mero de neur√¥nios na camada (aplic√°vel a camadas como `Dense`, `LSTM`, `GRU`, `SimpleRNN`).
- `kernel_size`: O tamanho do kernel usado nas opera√ß√µes de convolu√ß√£o (para `Conv2D`).
- `strides`: O stride da convolu√ß√£o (para `Conv2D`, `MaxPooling2D`).
- `padding`: O tipo de padding usado, `"valid"` ou `"same"` (para `Conv2D`, `MaxPooling2D`).
- `rate`: A taxa de dropout (para `Dropout`).
- `filters`: O n√∫mero de filtros de convolu√ß√£o (para `Conv2D`).

Estes s√£o apenas alguns dos hiperpar√¢metros. A escolha e o ajuste fino desses par√¢metros dependem da natureza espec√≠fica do problema e do conjunto de dados.
